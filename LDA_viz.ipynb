## Visualising Banking Inquiry Topic Clusters with pyLDAvis

Using the pyLDAvis tool, and following the steps outlined in this [notebook](http://nbviewer.jupyter.org/github/bmabey/pyLDAvis/blob/master/notebooks/pyLDAvis_overview.ipynb)

```python
import gensim
import pyLDAvis
```

Loading dictionary, corpus and model produced earlier

```python
dictionary = gensim.corpora.Dictionary.load('banking.dict')
corpus = gensim.corpora.MmCorpus('banking_corpus.mm')
lda = gensim.models.ldamodel.LdaModel.load('banking_lda.lda')
```

Instruct ipython to render the visualisation automatically

```python
pyLDAvis.enable_notebook()
```

Prepare the data for visualisation

```python
banking_vis_data = pyLDAvis.gensim.prepare(lda, corpus, dictionary)
```

### And tada!

Actually, there's a bit much overlap of terms to be able to clearly differentiate the clusters. That's probably due to the focused nature of the banking inquiry's work (it was established to investigate the causes of Ireland's economic crisis). Setting the relevance slider at 0.5-0.6, to give a greater weighting to topic specific frequency of terms as opposed to overall frequency, seems to produce more discernable clusters. 

If you can't see a chart below, check out the [browser version](banking_lda.html#topic=&lambda=0.6&term=)
